```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
```

```python
personality_scores = pd.read_csv('personality_scores.csv', sep=';')
departments = pd.read_csv('departments.csv', sep = ';')
display(personality_scores)
```

```python
personality_scores['ID'].duplicated().sum()
```

```python
personality_newdata = personality_scores.dropna(axis=1, how='all')
```

```python
personality_newdata['ID'].nunique() == personality_scores['ID'].nunique()
```

```python
personality_newdata.columns=personality_newdata.columns.str.replace('Section 5 of 6','')
personality_newdata.columns=personality_newdata.columns.str.replace('[','')
personality_newdata.columns=personality_newdata.columns.str.replace(']','')
```

```python
display(personality_newdata)
```

```python
catogories = {
        1: {
            "trait": "Extraversion",
            "key": 1
           },
        2: {
            "trait":"Agreeableness",
            "key": 2
            },
        3: {
            "trait":"Conscientiousness",
            "key": 3
        },
        4: {
            "trait":"Neuroticism",
            "key": 4
        },
        5: {
            "trait":"Intellect/Imagination",
            "key": 5
            }
        }

def sum_personality(item, key, data):
    count = 0
    for column in data.columns:
        score = item[column]
        if type(score) == str:
            result = int(score[4])
            if int(score[1]) == catogories[key]["key"]:
                count += result
    return count
```

```python
for cat in catogories:

    personality_newdata[catogories[cat]['trait']] = personality_newdata.apply(lambda row: sum_personality(row, cat, personality_newdata), axis=1)
```

```python
display(personality_newdata)
```
